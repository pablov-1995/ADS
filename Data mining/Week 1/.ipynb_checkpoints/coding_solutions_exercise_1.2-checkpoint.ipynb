```python
from collections import Counter
import spacy
import pandas as pd

from tqdm import tqdm_notebook
from scipy.stats import chi2_contingency

nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_pickle('discussions.p')
```

```python
posts = df.post.values
processed_texts = [text for text in tqdm_notebook(nlp.pipe(posts, 
                                              n_process=-1, 
                                              disable=["ner",
                                                       "parser"]),
                                          total=len(posts
                                                   ))]
```

```python
df['processed_texts'] = processed_texts
```

## 1.3

```python
flatten = lambda t: [item for sublist in t for item in sublist]

subset_1 = df[(df.title == 'Breaking Bad')].processed_texts
subset_1 = [[word.lemma_.lower() for word in text if word.pos_ == 'NOUN'] for text in subset_1] # i choose only nouns
subset_1 = flatten(subset_1)

subset_2 = df[(df.title == 'Better Call Saul')].processed_texts
subset_2 = [[word.lemma_.lower() for word in text if word.pos_ == 'NOUN'] for text in subset_2]
subset_2 = flatten(subset_2)
```

### Word that will differ

```python
counts_c1 = Counter(subset_1)
counts_c2 = Counter(subset_2)
```

```python
word2count = 'meth'
freq_c1 = counts_c1[word2count]
freq_c2 = counts_c2[word2count]

freq_c1_other = sum(counts_c1.values()) - freq_c1
freq_c2_other = sum(counts_c2.values()) - freq_c2

llr, p_value,_,_ = chi2_contingency([[freq_c1, freq_c2], 
                  [freq_c1_other, freq_c2_other]],
                  lambda_='log-likelihood') 
print("Log-likelihood: ", llr)
print('p-value', p_value)
```

### A word that doesn't

```python
word2count = 'television'
freq_c1 = counts_c1[word2count]
freq_c2 = counts_c2[word2count]

freq_c1_other = sum(counts_c1.values()) - freq_c1
freq_c2_other = sum(counts_c2.values()) - freq_c2

llr, p_value,_,_ = chi2_contingency([[freq_c1, freq_c2], 
                  [freq_c2_other, freq_c2_other]],
                  lambda_='log-likelihood') 

print("Log-likelihood: ", llr)
print('p-value', p_value)
```

## 1.4

```python
def distinctive_words(target_corpus, reference_corpus):
    counts_c1 = Counter(target_corpus) # don't forget to flatten your texts!
    counts_c2 = Counter(reference_corpus)
    vocabulary = set(list(counts_c1.keys()) + list(counts_c2.keys()))
    freq_c1_total = sum(counts_c1.values()) 
    freq_c2_total = sum(counts_c2.values()) 
    results = []
    for word in vocabulary:
        freq_c1 = counts_c1[word]
        freq_c2 = counts_c2[word]
        freq_c1_other = freq_c1_total - freq_c1
        freq_c2_other = freq_c2_total - freq_c2
        llr, p_value,_,_ = chi2_contingency([[freq_c1, freq_c2], 
                      [freq_c1_other, freq_c2_other]],
                      lambda_='log-likelihood') 
        if freq_c2 / freq_c2_other > freq_c1 / freq_c1_other:
            llr = -llr
        result = {'word':word, 
                    'llr':llr,
                    'p_value': p_value}
        results.append(result)
    results_df = pd.DataFrame(results)
    return results_df
```

```python
results_comp = distinctive_words(subset_1, subset_2)
```

```python
# distinctive for BB
results_comp.sort_values('llr', ascending=False)
```

```python
# distinctive for BCS
results_comp.sort_values('llr', ascending=True)
```

## Bonus points: duplicate distinctive words and vizualization

```python
results_df_bb['type'] = 'Breaking Bad'
results_df_bcs['type'] = 'Better Call Saul'
results_df_both['type'] = 'Both'
```

```python
results_all = pd.concat([results_df_bb, results_df_bcs, results_df_both])
results_all = results_all.sort_values('llr', ascending=False).drop_duplicates(['word'])
```

```python
top_20_both = list(results_all[results_all.type == 'Both'].word.values[:20])
top_20_bcs = list(results_all[results_all.type == 'Better Call Saul'].word.values[:20])
top_20_bb = list(results_all[results_all.type == 'Breaking Bad'].word.values[:20])
```

```python
set_1 = top_20_both + top_20_bcs
set_2 = top_20_both + top_20_bb 
```

```python
word2freq = dict(zip(results_all.word, results_all.llr))
```

```python
from matplotlib_venn_wordcloud import venn2_wordcloud

import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'DIN Alternate'
%matplotlib inline
fig, ax = plt.subplots(figsize=(10,10), dpi=300)

v = venn2_wordcloud([set(set_1), set(set_2)], 
                    ax=ax,
                    set_labels=['Better Call Saul', 'Breaking Bad'],
                   word_to_frequency=word2freq,
                    wordcloud_kwargs={'font_path': '/System/Library/Fonts/Supplemental/DIN Alternate Bold.ttf',
                                     'color_func':lambda *args, **kwargs: (0,0,0)})

v.get_patch_by_id('10').set_color('#27ae60')
v.get_patch_by_id('01').set_color('#f1c40f')
v.get_patch_by_id('11').set_color('#3498db')
```
